```python
import matplotlib.pyplot as plt
import pandas as pd
# plot MountainCarContinuous

fig, ax1 = plt.subplots(1, 1)    # a figure with a 2x1 grid of Axes
ddpg_m = pd.read_csv("../exps/expDDPG_Mount.csv")
td3_m = pd.read_csv("../exps/expTD3_Mount.csv")
sac_m = pd.read_csv("../exps/expSAC_Mount.csv")

ax1.plot(ddpg_m['Step'], ddpg_m['Value'], color="violet", label='DDPG')
ax1.plot(td3_m['Step'], td3_m['Value'], color="blue", label='TD3')
ax1.plot(sac_m['Step'], sac_m['Value'], color="green", label='SAC')


#ax1.set_xticks(np.arange(0, 24, step=2))
ax1.set_xlabel("timesteps")
ax1.set_ylabel("Running Reward")
ax1.set_title("Running Reward in MountainCarContinuous")
plt.show()
fig.savefig(fname='./'+'MountainCarContinuous.png', format='png')
```

```python
import plotly.graph_objects as go
import pandas as pd

fig = go.Figure()

ddpg_m = pd.read_csv("../exps/expDDPG_Mount.csv")
td3_m = pd.read_csv("../exps/expTD3_Mount.csv")
sac_m = pd.read_csv("../exps/expSAC_Mount.csv")

fig.add_trace(go.Scatter(x=ddpg_m['Step'], y=ddpg_m['Value'], mode='lines', name='DDPG'))
fig.add_trace(go.Scatter(x=td3_m['Step'], y=td3_m['Value'], mode='lines', name='TD3'))
fig.add_trace(go.Scatter(x=sac_m['Step'], y=sac_m['Value'], mode='lines', name='SAC'))

fig.show()
fig.write_image('./'+'MountainCarContinuous.png')
```
